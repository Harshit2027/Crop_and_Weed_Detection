# Import Libraries

```python
import os
import cv2
import numpy as np
from keras.models import load_model, Model
from keras.optimizers import Adam
from keras.layers import Dense, Conv2D
from keras.utils import to_categorical
```

# Creating custom model from YOLOV3 model

```python
def create_custom_model(pretrained_model, num_classes, num_boxes):
    # Freeze the layers of the pre-trained model
    for layer in pretrained_model.layers:
        layer.trainable = False

    # Add a custom classification head for your two classes
    x = pretrained_model.layers[-2].output

    # Assuming num_boxes is the number of anchor boxes, and num_classes is the number of classes
    output_layer = Conv2D(num_boxes * (4 + num_classes), (1, 1), activation='linear')(x)

    model = Model(inputs=pretrained_model.input, outputs=output_layer)

    return model
```

# Loading data from directory

```python
def load_data(image_folder, annotation_folder):
    image_paths = []  # List to store image file paths
    annotations = {}  # Dictionary to store annotations

    # Load image paths
    for image_file in os.listdir(image_folder):
        image_paths.append(os.path.join(image_folder, image_file))

    # Load annotations from YOLO format annotation files
    for annotation_file in os.listdir(annotation_folder):
        with open(os.path.join(annotation_folder, annotation_file), 'r') as f:
            lines = f.readlines()
            annotation_info = []
            for line in lines:
                line = line.strip().split()
                class_index = int(line[0])
                x_center = float(line[1])
                y_center = float(line[2])
                width = float(line[3])
                height = float(line[4])
                annotation_info.append((class_index, x_center, y_center, width, height))
            annotations[annotation_file.replace('.txt', '.jpg')] = annotation_info

    return image_paths, annotations
```

# Proceeding Normalization

```python
def convert_annotations_to_labels(annotations, num_classes):
    labels = []
    grid_size = 32
    
    for annotation_info in annotations.values():
        label = np.zeros((32, 32, 2))  # Initialize label with zeros

        for annotation in annotation_info:
            class_index = annotation[0]
            x_center = annotation[1]
            y_center = annotation[2]
            width = annotation[3]
            height = annotation[4]

            # Assuming the bounding box coordinates are normalized to image size (0 to 1)
            x_min = int(x_center * 32)
            y_min = int(y_center * 32)
            x_max = int((x_center + width) * 32)
            y_max = int((y_center + height) * 32)

            label[y_min:y_max, x_min:x_max, :] = to_categorical(class_index, num_classes=num_classes)

        labels.append(label)

    return np.array(labels)
```

# Training our model

```python
def train_custom_model(custom_model, image_paths, annotations, num_classes):
    
    # Convert annotations to labels
    labels = convert_annotations_to_labels(annotations, num_classes)

    # Convert image paths to NumPy array (you may need to resize and preprocess the images here)
    images = []
     
    for image_path in image_paths:
        image = cv2.imread(image_path)
        # Resize the image to the target size
        image_resized = cv2.resize(image, (512, 512))
        # Perform any additional preprocessing steps here, e.g., normalization
        # For example, if you want to normalize pixel values to the range [0, 1]
        image_preprocessed = image_resized / 255.0
        images.append(image_preprocessed)

    images = np.array(images)
    
    # images = np.array([cv2.imread(image_path) for image_path in image_paths])

    # Split the data into training and validation sets (80% training, 20% validation)
    split_idx = int(0.8 * len(images))
    x_train, x_val = images[:split_idx], images[split_idx:]
    y_train, y_val = labels[:split_idx], labels[split_idx:]

    # Compile the custom model with appropriate optimizer, loss, and metrics
    custom_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    custom_model.fit(x_train, y_train, batch_size=16, epochs=10, validation_data=(x_val, y_val))
```

# Bounding box function for detecting objects

```python
def detect_objects(model, image_path, class_names):
    # Load the image
    image = cv2.imread(image_path)

    # Preprocess the image (resize and normalize)
    image_resized = cv2.resize(image, (512, 512))
    image_preprocessed = image_resized / 255.0

    # Add a batch dimension to the image
    image_input = np.expand_dims(image_preprocessed, axis=0)

    # Use the model to predict the bounding boxes and classes
    predictions = model.predict(image_input)
    boxes = predictions[0]  # Get the predicted bounding boxes for the first image in the batch
                           # Assuming predictions have shape (1, grid_size, grid_size, num_classes + 4)

        
    # Get the dimensions of the input image (img_height, img_width)
    img_height, img_width, _ = image.shape

    # Process the predictions and draw bounding boxes
    for y in range(32):
        for x in range(32):
            box = boxes[y, x]  # Get the predictions for the current cell

            # Extract class probabilities and box coordinates
            class_probabilities = box[:num_classes]
            class_index = np.argmax(class_probabilities)
            confidence = class_probabilities[class_index]

            # Check if the confidence for the predicted class is high enough
            if confidence > 0.5:
                class_name = class_names[class_index]
                
                # Extract box coordinates
                x_min, y_min, x_max, y_max = box[num_classes:num_classes+4]
                
#                 # Print the bounding box coordinates
#                 print(f"Class: {class_name}, Confidence: {confidence:.2f}, "
#                       f"Bounding Box Coordinates: ({x_min:.2f}, {y_min:.2f}, {x_max:.2f}, {y_max:.2f})")

                
                x_min = int(x_min * img_width)  # Convert normalized coordinates to pixels
                y_min = int(y_min * img_height)
                x_max = int(x_max * img_width)
                y_max = int(y_max * img_height)
                

                # Draw the bounding box on the image
                cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                cv2.putText(image, class_name, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    return image


#2nd method

# def detect_objects(model, image_path, class_names):
#     image = cv2.imread(image_path)
#     image_copy = image.copy()  # Create a copy of the original image
#     image_copy = cv2.resize(image_copy, (416, 416))
#     image_copy = image_copy / 255.0
#     image_copy = np.expand_dims(image_copy, axis=0)

#     detections = model.predict(image_copy)

#     # Process detections and draw bounding boxes on the image
#     grid_size = detections.shape[1]  # Assuming grid_size x grid_size grid for YOLO-like models

#     for grid_row in range(grid_size):
#         for grid_col in range(grid_size):
#             for anchor_idx in range(3):  # Assuming 3 anchor boxes for YOLO-like models
#                 object_info = detections[0, grid_row, grid_col, anchor_idx * (num_classes + 5): (anchor_idx + 1) * (num_classes + 5)]
#                 class_probs = object_info[5:]  # Class probabilities

#                 if np.any(class_probs):  # Check if class_probs is not empty
#                     class_index = np.argmax(class_probs)  # Get the class index with the highest probability
#                     confidence = object_info[4]  # Confidence score

#                     if confidence > 0.5:  # You can adjust this threshold as needed
#                         # Calculate the bounding box coordinates
#                         x_center = (object_info[0] + grid_col) / grid_size
#                         y_center = (object_info[1] + grid_row) / grid_size
#                         width = object_info[2]
#                         height = object_info[3]

#                         x_min = int((x_center - width / 2) * 416)
#                         y_min = int((y_center - height / 2) * 416)
#                         x_max = int((x_center + width / 2) * 416)
#                         y_max = int((y_center + height / 2) * 416)

#                         class_name = class_names[class_index]

#                         # Draw bounding box on the image_copy (not on the original image)
#                         cv2.rectangle(image_copy, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)  # Green rectangle
#                         text = f"{class_name} ({confidence:.2f})"
#                         cv2.putText(image_copy, text, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

#     return image_copy

```
